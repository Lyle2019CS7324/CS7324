## 2. Data understanding

### 2.1 Data cleanning

```python
#attribute pandas & numpy to csv.
import pandas as pd
import numpy as np

print('Pandas:', pd.__version__)
print('Numpy:',np.__version__)

#import raw data, named as 'cars'
cars = pd.read_csv('https://raw.githubusercontent.com/sebastian0303/CS7324/master/CS%207324%20LAB%20ONE%20.csv')

#TEST READING
cars.head()
```

```python
print(cars.info())
```

```python
# capitalize columns name to avoid colliding reserved keywords
cars.columns = map(str.capitalize, cars.columns)
```

```python
import matplotlib
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore', DeprecationWarning)
%matplotlib inline 

import missingno as mn

# clean the dataset

for col in ['Url','Vin', 'Lat', 'Long', 'County_fips', 'County_name', 'State_code',  'City', 'Weather', 'Make']:
    if col in cars:
        del cars[col]

cars_imputing = cars.dropna(subset=['Year','Type','Condition','Manufacturer','Cylinders','Drive','Paint_color','Size'])

mn.matrix(cars)
```

```python
# group attributes relevant to odometer
cars_grouped = cars_imputing.groupby(by=['Year','Condition','Type'])

cars_imputed = cars_grouped.transform(lambda grp: grp.fillna(grp.median()))

col_deleted = list( set(cars_imputing.columns) - set(cars_imputed.columns))
cars_imputed[col_deleted] = cars_imputing[col_deleted]

cars_imputed.dropna(inplace=True)

print(cars_imputed.info())
```

```python
# convert year to integer
cars_imputed = cars_imputed.astype({'Year': 'int64'})
# quantification
cars_imputed.Cylinders.replace(to_replace=['6 cylinders','5 cylinders','4 cylinders','other','10 cylinders','12 cylinders','3 cylinders','8 cylinders'],
                               value=[6,5,4,0,10,12,3,8],
                               inplace=True)

cars_imputed.Condition.replace(to_replace=['salvage', 'fair', 'good', 'excellent', 'like new', 'new'],
                               value=[1,2,3,4,5,6],
                               inplace=True)

cars_imputed.Size.replace(to_replace=['sub-compact', 'compact', 'mid-size', 'full-size'],
                               value=[1,2,3,4],
                               inplace=True)

cars_imputed.info()
```

```python
# fake ads put price odometer on 1 to get a very top sorting
cars_imputed = cars_imputed[cars_imputed.Price != 1]
cars_imputed = cars_imputed[cars_imputed.Odometer != 1]

# impute noise < Q1 - 1.5*IQE and noise > Q3 + 1.5*IQE
Q1 = cars_imputed.quantile(0.25)
Q3 = cars_imputed.quantile(0.75)
IQR = Q3 - Q1

# print(Q1,"\n===========\n", Q3)

noise_odometer = (cars_imputed.Odometer < (Q1.Odometer - 1.5 * IQR.Odometer)) | (cars_imputed.Odometer > (Q3.Odometer + 1.5 * IQR.Odometer))
noise_price = (cars_imputed.Price < (Q1.Price - 1.5 * IQR.Price)) | (cars_imputed.Price > (Q3.Price + 1.5 * IQR.Price))

cars_imputed = cars_imputed[~(noise_odometer | noise_price)]

cars_imputed.info()
```

## 3. Data visualization 

```python
import matplotlib
import matplotlib.pyplot as plt
import warnings
import seaborn as sns

warnings.simplefilter('ignore', DeprecationWarning)
%matplotlib inline 

print('Matplotlib:', matplotlib. __version__)

cmap = sns.diverging_palette(220, 10, as_cmap=True) # one of the many color mappings

print('Seaborn:', sns. __version__)

plt.style.use('ggplot')
```

```python
# break up year variable, hiztorical cars are not our target
# https://dmv.dc.gov/service/historical-vehicles
# Historical license plates are new plates issued by the DMV to a vehicle that was manufactured more than 25 years before the current calendar year that is used only as a collector's item or exhibition piece, and not for daily transportation.
cars_imputed['Year_range'] = pd.cut(cars_imputed['Year'],[1920, 1994, 2019],
                                   labels=['Historical', 'Used'])

used_cars = cars_imputed[cars_imputed.Year_range=="Used"]
used_cars.describe()

cars_imputed.Year_range.describe()
```

```python
# distribution of manufacturers and Type
categories = pd.crosstab(used_cars.Manufacturer,used_cars.Type)

categories['Sum']=categories.sum(axis=1)

categories.sort_values('Sum', ascending=False, inplace=True)
categories.drop('Sum',axis=1, inplace=True)

categories.plot(kind='bar', stacked=True, figsize=(15,5))
plt.show()
```

```python
# check Title and location distribution
titles = pd.crosstab(used_cars.State_name,used_cars.Title_status)

titles.plot(kind='bar', stacked=True, figsize=(15,5))
plt.show()
```

```python
# check drive and color distribution
titles = pd.crosstab(used_cars.Paint_color,used_cars.Drive)

titles.plot(kind='bar', stacked=True, figsize=(15,5))
plt.show()
```

```python
# fuel
used_cars.groupby('Fuel').Year.count().plot.bar()
plt.show()
```

```python
fig, ax = plt.subplots(4, figsize=(10,16))

target_cars = used_cars[(used_cars['Type'].isin(target_types))]
target_cars = target_cars[(target_cars['Manufacturer'].isin(target_brands))]

target_grouped = target_cars.groupby(['Type'])

for i, (val, grp) in enumerate(target_grouped):
    sns.lineplot(x="Year", y="Price", hue="Manufacturer", data=grp, ci="sd",
                 estimator="median", ax=ax[i]).set_title(val + " price distribution with SD")
plt.tight_layout()
plt.show()
```
